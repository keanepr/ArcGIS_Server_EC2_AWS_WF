### Select prefire and midfire scenes from the image collection

Here we select pre-fire and mid-fire satellite scenes. We define an area of interest and start off by first running a preliminary search for all applicable scenes between two approximate dates.

```python
aoi = {'spatialReference': {'latestWkid': 3857, 'wkid': 102100},
       'xmax': -13643017.100720055,
       'xmin': -13652113.10708598,
       'ymax': 4739654.477447927,
       'ymin': 4731284.622850712}
arcgis.env.analysis_extent = aoi
sentinel.extent = aoi
```

In this particular image collection, there are only two scenes, but if we have more scenes in the collection, to limit our search results to a particular date range and area of interest, we can optionally use `filter_by()` by specifying the where and geometry parameter.

```python
selected = sentinel.filter_by(where="acquisitiondate BETWEEN timestamp '2018-06-15 00:00:00' AND timestamp '2018-06-24 19:59:59'",
                              geometry=arcgis.geometry.filters.intersects(aoi))
```

## Quantitative Assessment

The **Normalized Burn Ratio (NBR)** can be used to delineate the burnt areas and identify the severity of the fire. 

The formula for the NBR is very similar to that of the Normalized Difference Vegetation Index (NDVI), except that it uses near-infrared band 9 and the short-wave infrared band 13:
\begin{align}
{\mathbf{NBR}} = \frac{\mathbf{B9} - \mathbf{B13}}{\mathbf{B9} + \mathbf{B13} + \mathbf{WS}} \\   
\end{align}

The NBR equation was designed to be calculated from reflectance, but it can be calculated from the radiance and digital_number_(dn) with changes to the burn severity table below. The WS parameter is used for water suppression and is typically 2000. 

For a given area, NBR is calculated from an image just prior to the burn and a second NBR is calculated for an image immediately following the burn. Burn extent and severity are judged by taking the difference between these two index layers:

\begin{align}
{\Delta \mathbf{NBR}} = \mathbf{NBR_{prefire}} - \mathbf{NBR_{postfire}} \\   
\end{align}

The meaning of the âˆ†NBR values can vary by scene, and for best results, interpretation in specific instances should always be based on some field assessment. However, the following table from the USGS FireMon program can be useful as a first approximation for interpreting the NBR difference:


| \begin{align}{\Delta \mathbf{NBR}}  \end{align}      | Burn Severity |
| ------------- |:-------------:|
| 0.1 to 0.27   | Low severity burn |
| 0.27 to 0.44  | Medium severity burn |
| 0.44 to 0.66 | Moderate severity burn |
| > 0.66 | High severity burn |

[ Source: http://gsp.humboldt.edu/OLM/Courses/GSP_216_Online/lesson5-1/NBR.html ]

### Use Band Arithmetic and Map Algebra 

```python
nbr_prefire = band_arithmetic(prefire, "(b9 - b13) / (b9 + b13 + 2000)")
nbr_postfire = band_arithmetic(midfire, "(b9 - b13) / (b9 + b13 + 2000)")

nbr_diff = nbr_prefire - nbr_postfire
```

```python
burnt_areas = colormap(remap(nbr_diff,
                             input_ranges=[0.1,  0.27,  # low severity
                                           0.27, 0.44,  # medium severity
                                           0.44, 0.66,  # moderate severity
                                           0.66, 1.00],  # high severity burn
                             output_values=[1, 2, 3, 4],
                             no_data_ranges=[-1, 0.1], astype='u8'),
                       colormap=[[4, 0xFF, 0xC3, 0], [3, 0xFA, 0x8E, 0], [2, 0xF2, 0x55, 0], [1, 0xE6, 0,    0]])
```

```python
# Visualize burnt areas
burnt_areas
```

With that, we computed the NBR on scenes from before and during the burn and computed the NBR difference to identify places that had been affected by the fire. We've also normalized the values to match a burn severity index and applied a color map that brings out the extent of the damage caused by the fire.

### Area calculation

A histogram shows us how the data are distributed. Here we use `compute_histogram()` to plot the severity distribution and to calculate the burnt area.

As the histogram result contains information about the frequency of pixels that fall within each category of severity, performing a sum operation on these will give you the total number of pixels. Multiplying this with the pixel size will give you the total area.

The pixel size along the x and y direction can be calculated by performing (extent along the x-direction/number of columns) and (extent along the y-direction/number of rows), respectively. 

You may specify any value for the number of columns and rows for pixel size calculation, but the same pixel size would have to be used while computing the area in the next step (`sqmarea = numpix * pixx * pixy`). Here, 1200 is used as the number of columns and 450 as the number of rows, as these are the default values used for displaying imagery layers.

```python
pixx = (aoi['xmax'] - aoi['xmin']) / 1200.0
pixy = (aoi['ymax'] - aoi['ymin']) / 450.0

res = burnt_areas.compute_histograms(aoi, pixel_size={'x': pixx, 'y': pixy})

numpix = 0
histogram = res['histograms'][0]['counts'][1:]
for i in histogram:
    numpix += i
```

### Report burnt area

```python
sqmarea = numpix * pixx * pixy  # in sq. m
acres = 0.00024711 * sqmarea   # in acres

HTML('<h3>Fire has consumed <font color="red">{:,} acres</font>  till {}</h3>.'
     .format(int(acres), df.iloc[-1]['AcquisitionDate'].date()))
```

```python
%matplotlib inline

plt.title('Distribution by severity', y=-0.1)
plt.pie(histogram, labels=[
        'Low Severity', 'Medium Severity', 'Moderate Severity', 'High Severity'])
plt.axis('equal')
```

## Raster to Feature layer conversion

Use Raster Analytics and Geoanalytics to convert the burnt area raster to a feature layer. The `to_features()` method converts the raster to a feature layer, and `create_buffers()` fills holes in the features and dissolves them to output one feature that covers the extent of the Pawnee Fire.

```python
burnt_areas = burnt_areas.layers[0]
fire_item = burnt_areas.to_features(
    output_name=f'Pawnee_Fire_Feature_Layer_{timestamp}', gis=gis)
fire_layer = fire_item.layers[0]
```

```python
fire = create_buffers(fire_layer, 100, 'Meters', dissolve_option='All',
                      multipart=True, output_name=f'pawnee_fire_area_{timestamp}')
fire = fire.layers[0]
```

```python
data_groups = gis.groups.search('"Esri Sample Notebooks Data" owner:esri_notebook',
                                outside_org=True)
group_query = f"group: {data_groups[0].id}" if data_groups else ""
fire = gis.content.search(f"title: pawnee_fire_area {group_query}",
                          outside_org=True)[0]
print(fire)
```
